```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
 
Temperature = [0, 0.3, 0.5, 0.7, 0.9, 1.0]
Accuracy = [0.7680412371134021,
          0.7577319587628866,
         0.7422680412371134,
         0.7061855670103093,
          0.5257731958762887,
          0.520618556701031]
 
data_plot = pd.DataFrame({"Temperature":Temperature, "Accuracy":Accuracy})
 
sns.lineplot(x = "Temperature", y = "Accuracy", data=data_plot)
plt.title('Effect of temperature on unit testing metric')
plt.show()
```
